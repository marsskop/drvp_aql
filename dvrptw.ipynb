First, we create the solution state.

```python
# solution class
class DVRPTWState:
    """
    Solution state for DVRP.
    Attributes:
        routes: A list of list of integers, where each inner list corresponds to a single route denoting the sequence of customers to be visited. A route does not contain the depot.
        removal: A list of integers, each integer representing a customer in a removal list.
    """
    def __init__(self, routes, removal=None):
        self.routes = routes
        self.removal = removal if removal is not None else []
    
    def copy(self):
        return DVRPTWState(copy.deepcopy(self.routes), self.removal.copy())
    
    def objective(self):
        """
        Computes the total route cost.
        """
        return sum(route_cost(route) for route in self.routes)
    
    @property
    def cost(self):
        """
        Alias for objective method. Used for plotting.
        """
        return self.objective()
    
    def find_route(self, customer):
        """
        Return the route that contains the passed-in customer.
        """
        for route in self.routes:
            if customer in route:
                return route
        raise ValueError(f"Solution does not contain customer {customer}")


def distance(node1, node2):
    """
    Calculates distances between nodes.
    """
    return np.sqrt((node1["XCOORD"].item() - node2["XCOORD"].item()) ** 2 + (node1["YCOORD"].item() - node2["YCOORD"].item()) ** 2)


def route_cost(route):
    """
    Calculates cost of the route, taking the DEPOT in consideration. If sum of demand enroute is bigger than VEHICLE_CAPACITY, returns 1e6.
    """
    tour = [DEPOT] + [df[df.CUST == x] for x in route] + [DEPOT]
    cost = sum(distance(tour[i], tour[i+1]) for i in range(len(tour) - 1))
    demand = sum(df[df.CUST == x].DEMAND.item() for x in route)
    if demand > VEHICLE_CAPACITY:
        return 1e6
    return cost
```

```python
def plot_solution(solution, name="DVRPTW Solution"):
    """
    Plot the routes of the passed-in solution.
    """
    fig, ax = plt.subplots()
    cmap = matplotlib.cm.rainbow(np.linspace(0, 1, len(solution.routes)))

    for idx, route in enumerate(solution.routes):
        ax.plot(
            [0] + [df[df.CUST == loc].XCOORD.item() for loc in route] + [0],
            [0] + [df[df.CUST == loc].YCOORD.item() for loc in route] + [0],
            color=cmap[idx],
            marker='.'
        )

    ax.scatter(DEPOT["XCOORD"], DEPOT["YCOORD"], c="tab:red", label="Depot", zorder=3, marker="*", s=750)


    ax.set_title(f"{name}\nTotal distance: {solution.cost}\nNumber of routes: {len(solution.routes)}")
    ax.set_xlabel("X-coordinate")
    ax.set_ylabel("Y-coordinate")
    ax.legend(frameon=False, ncol=3)
```

Next, let's specify the initial solution that builds a solution out of static customers. This solution will be repeatedly destroyed and repaired by the ALNS heuristics.

The algorithm starts with an empty solution and then iteratively adds customers with the least cost to the routes. If there are no routes available, then a new route is added. This algoirthm is simlar to the Nearest Neighbor heuristic.

```python
MU = 1  # a coefficient that controls the relative importance of the travel distance and time constraint
def feasibility_cost(k, i, j):
    """
    The feasibility cost of inserting customer k between customers i and j. If customer k can't be inserted between i and j, return None.
    """
    k_cust = df[df.CUST == k]
    i_cust = df[df.CUST == i]
    j_cust = df[df.CUST == j]
    if i == 0:
        i_cust = DEPOT
    if j == 0:
        j_cust = DEPOT
    
    # Check time feasibility
    if i_cust.READYTIME.item() > k_cust.READYTIME.item() and i_cust.DUEDATE.item() >= k_cust.DUEDATE.item():
        return None
    if (j_cust.READYTIME.item() < k_cust.READYTIME.item() and j_cust.DUEDATE.item() <= k_cust.DUEDATE.item()) and j_cust.READYTIME.item() > 0:
        return None

    # Extra travel distance caused by insertion of k
    d_ikj = distance(i_cust, k_cust) + distance(k_cust, j_cust) - distance(i_cust, j_cust)
    # Delay of the service start time of j after inserting k
    t_ikj = 0  # not sure, how to calculate the delay: b' - b in the article, what is b'?
    return MU * d_ikj + (1 - MU) * t_ikj

def least_cost(removal, route):
    """
    Return the customers and their best positions with the least cost of insertion into the route, excluding the depot.
    """
    cust_pos_cost = []  # in form of [(k, i, cost), ...] where k is the customer to insert, i is the index of the customer in route after which to insert, cost is the cost of insertion
    for cust in removal:
        # For each customer calculate feasibility cost of insertion between adjacent customers i and i+1 in the route (on position i+1) + edge
        costs = [feasibility_cost(cust, route[i], route[i+1]) for i in range(len(route)-1)] + [feasibility_cost(cust, route[-1], 0)]
        costs = [cost for cost in costs if cost is not None]
        i = np.argmin(costs)
        min_cost = costs[i]
        cust_pos_cost.append((cust, i+1, min_cost))  # on position i+1!
    return sorted(cust_pos_cost, reverse=True, key=lambda x: x[2])

def seed(removal, type="farthest"):
    """
    Return the seed customer from removal list who is farthest from the depot or has the earliest deadline of service.
    """
    if type == "farthest":
        return max(removal, key=lambda x: distance(DEPOT, df[df.CUST == x]))
    elif type == "earliest":
        return min(removal, key=lambda x: df[df.CUST == x].DUEDATE.item())
    else:
        raise ValueError(f"Wrong type for seed customer evaluation: {type}")

def init_sol():
    """
    Build initial solution by iteratively constructing routes, where the customer with the least cost is added until the route has met the vehicle capacity limit.
    """
    routes = []
    removal = set(df["CUST"])

    while removal:
        # Start at the depot
        route = [0]
        route_demands = 0

        # And add seed customer
        seed_cust = seed(removal)
        route.append(seed_cust)
        route_demands += df[df.CUST == seed_cust].DEMAND.item()
        removal.remove(seed_cust)

        while removal:
            # Add the least costly customer to the route until max capacity
            nearest, nearest_pos, _ = least_cost(removal, route)[0]

            if route_demands + df[df.CUST == nearest].DEMAND.item() > VEHICLE_CAPACITY:
                break

            route.insert(nearest_pos, nearest)
            removal.remove(nearest)
            route_demands += df[df.CUST == nearest].DEMAND.item()
        
        customers = route[1:]  # remove the depot
        routes.append(customers)
    
    return DVRPTWState(routes)
```

# Destroy Solution

```python
from functools import reduce
```

Implements three destroy heuristics.

1. Random Removal (RR). The customers are randomly selected for remove operation. This heuristic helps diversify the search and escape local optima.

2. Worst Removal (WR). The customers are removed if deleting them leads to maximum cost savings. For a customer $k$ the index $J_k$ is defined as follows, where $F_{-k}(S)$ is the cost of solution without customer $k$.:
$$ J_k = F(S) - F_{-k}(S) $$

3. Static-related Removal (SR). This heuristic attempts to remove the customers that are similar in some manner. We randomly select a reference customer $i$ from removal, then we calculate the index as the relevance of the reference customer $i$ to each customer in routes. The measure is as follows, where $M_d$ and $M_t$ are scaling constants, $\alpha$ and $\beta$ define the weight of relatedness based on route distance or time windows respectively.
$$ J_k = r_{ij} = \alpha * \frac{d_ij}{M_d}+ \beta * \frac{|l_i - l_j|}{M_t} $$

```python
def random_removal(solution, enroute, params):
    return [(c,v) for c,v in zip(list(enroute), np.random.random(size=len(enroute)))]

def worst_removal(solution, enroute, params):
    score = []
    for c in enroute:
        route = solution.find_route(c)
        route_ww = route[::]
        route_ww.remove(c)
        score.append((c, -(route_cost(route) - route_cost(route_ww))))
    return score

def static_rel_removal(solution, enroute, params):
    if solution.removal:
        ref = np.random.choice(list(solution.removal))
    else:
        ref = np.random.choice(list(enroute))
    
    node_ref = df[df.CUST == ref].iloc[0]
    score = []
    for c in enroute:
        node_c = df[df.CUST == c].iloc[0]
        D = distance(node_ref, node_c)
        T = np.abs(node_ref.DUEDATE - node_c.DUEDATE)
        score.append((c, params['alpha'] * D / params['M_D'] + params['beta'] * T / params['M_T']))
    return score


def time_rel_removal(solution, enroute, params):
    raise NotImplementedError("Нужно слишком много новых параметров")
```

```python
def destroy_solution(solution, destroy_score_fn, params):
    del_sol = solution.copy()
    del_sol.removal = set(del_sol.removal)
    enroute = reduce(lambda x, y: x | y, [set(r) for r in del_sol.routes], set()) # TODO: Add to solution class

    while len(del_sol.removal) < params['q']:
        scores = sorted(destroy_score_fn(del_sol, enroute, params), key=lambda x: x[1])
        z = np.random.random()
        i = int((z ** params['d']) * len(enroute))
        del_sol.removal.add(scores[i][0])
        del_sol.find_route(scores[i][0]).remove(scores[i][0])
        enroute.remove(scores[i][0])
    
    del_sol.removal = list(del_sol.removal)
    return remove_empty_routes(del_sol)

def remove_empty_routes(sol):
    sol.routes = [route for route in sol.routes if len(route) != 0]
    return sol
```

# Repair Solution

Implements three repair heuristics.

1. Greedy Insertion (GI). Repeatedly inserts customers according to a cost measure. The insert cost is calculated as follows, if a customer $i$ can be inserted into the route at position $k$, else $H(i, k, r) = \inf$.
$$\Delta_i = \Delta H(i, k, r) = d_{ik} + d_{k(i+1)} - d_{i(i+1)} $$
To determine the best insert position, we calculate $arg min H(i, k, r)$.

2. Regret Insertion (RI). Based on a regret value, or look-ahead value, which considers not only the best insert position but also the second, the third, etc. A regret-$k$ insertion locates where to put the customer with the biggest cost difference that is between the best placement and the $k-1$ alternatives. The regret value is calculated as follows, where $F_{k}(i)$ is the $k$th best place to insert customer $i$.
$$ \Delta_i = \max_{i \in L} \{ \sum_{h=2}^k (F_h(i) - F_1(i)) \} $$
The customer is inserted according to $i^* = arg \max_{i \in L} \Delta_i $. We are mainly concerned with two regret levels: regret-2 (R2I) and regret-3 (R3I).

```python
def feasibility_cost_demand(k, i, j, route):
    h = feasibility_cost(k, i, j)
    demand = sum(df[df.CUST == x].DEMAND.item() for x in route) + df[df.CUST == k].DEMAND.item()
    if demand > VEHICLE_CAPACITY:
        return None
    return h

def greedy_insertion(sol: DVRPTWState, rnd):
    """
    Greedy Insertion (GI) repair heuristic.
    """
    removal = set(sol.removal)
    rep_sol = sol.copy()
    while removal:
        best_cost = float('inf')
        best_cust, best_route, best_pos = None, None, None
        for cust in removal:
            for r_idx, route in enumerate(rep_sol.routes):
                for pos in range(len(route)):
                    # Calculate the cost of inserting the customer at the current position
                    delta_h = feasibility_cost_demand(cust, route[pos-1], route[pos], route)
                    if delta_h is not None and delta_h < best_cost:
                        best_cost = delta_h
                        best_cust, best_route, best_pos = cust, r_idx, pos
        if best_cust is not None:
            rep_sol.routes[best_route].insert(best_pos, best_cust)
            removal.remove(best_cust)
        else:
            rep_sol.routes.append([removal.pop()])
    rep_sol.removal = list(removal)
    return rep_sol


def regret_insertion(sol: DVRPTWState, rnd, k=2):
    """
    Regret Insertion (RI) repair heuristic.
    """
    removal = set(sol.removal)
    rep_sol = sol.copy()
    while removal:
        regret_values = []
        for cust in removal:
            best_positions = sorted(list(filter(lambda x: x[2] is not None,
                                    [(r_idx, pos, feasibility_cost_demand(cust, route[pos-1], route[pos], route))
                                    for r_idx, route in enumerate(rep_sol.routes)
                                    for pos in range(len(route))])),
                                    key=lambda x: x[2])[:k]
            if len(best_positions) >= 2:
                regret_value = sum(pos[2] for pos in best_positions[1:]) - best_positions[0][2]
                regret_values.append((cust, regret_value, best_positions[0]))

        if regret_values:
            max_regret_cust, _, (r_idx, pos, _) = max(regret_values, key=lambda x: x[1])
            rep_sol.routes[r_idx].insert(pos, max_regret_cust)
            removal.remove(max_regret_cust)
        else:
            rep_sol.routes.append([removal.pop()])
    rep_sol.removal = list(removal)
    return rep_sol
```

# ALNS

```python
def random_destroy(x, y):
    return destroy_solution(x, random_removal, params)

def worst_destroy(x, y):
    return destroy_solution(x, worst_removal, params)

def static_rel_destroy(x, y):
    return destroy_solution(x, static_rel_removal, params)

def regret_repair(x, y):
    return regret_insertion(x, None, k=3)
```

The ALNS algorithm is an extension of Large Neighborhood Search algorithm, based on destroy and repair heuristics. ALNS applies multiple competing destroy and repair heuristics that are applied according to their performance in the past.

- **Large neighborhood**. The neighborhood of a solution is enlarged by a destroy heuristic, which removes some customers from the solution, and a repair method, which rebuilds the partial solution, thereby improving the overall search. At each iteration step, some customers are removed from the current solution and placed in a removal list. Then, some customers are selected from the removal list to insert into the current partial solution using the repair heuristics. If there are some customers that cannot be inserted, their number will be used to increase the penalty in the objective function accordingly. If the solution is accepted by some criterion, the current iteration is over, and the remaining customers are used to set up the removal list for the next iteration.

We use `alns.ALNS` algorithm.

- **Adaptive search & Adaptive weight adjustment**. Several heuristics are considered during iteration. A heuristic (either the removal or the insertion heuristics) is selected by using a roulette-wheel selection principle where each of the heuristics is assigned a weight depending on its past performance. By combining the destroy and repair heuristics, we assign a weight to each pair, rather than rewarding each destroy and repair operator individually. The probability for choosing an operator pair is proportional to the weight $\omega_{dr}$ of each destroy-repair pair $(d, r)$ and, given a destroy and repair heuristics $\eta_d$ and $\eta_r$ , is as follows:
$$ \rho_{dr} = \frac{\omega_{dr}}{\sum_{d=1}^{\eta_d} \sum_{r=1}^{\eta_r} \omega_{dr}} $$
The weights of destroy-repair pairs are adjusted not every iteration, but for the length of the segment $\phi$.

We use `alns.select.SegmentedRouletteWheel` for heuristics selection.

- **Later perturbation**. After $\varphi$ repeat times, we apply a 2-opt* operator to each vehicle route with the aim of enhancing the improvement and modifying an infeasible solution to be feasible.

We use `two_opt_star` function for later perturbation.

- **Acceptance criterion**. After finding a new solution we check whether it can be accepted. The Record-to-Record Travel (RRT) criterion accepts a candidate solution if the absolute gap between the candidate and the best or current solution is smaller than a threshold. The threshold $T$ is updated in each iteration using a step size $\gamma$ as follows:
$$T \gets \max\{ T_{end}, \gamma T \}$$

We use `alns.accept.RecordToRecordTravel` for accept criterion.

- **Stop criterion**. The ALNS stops when a specific number of consecutive unimproved iterations $M$ have been run.

We use `alns.stop.MaxIterations` for stop criterion.

```python
import numpy.random as rnd
from alns import ALNS
from alns.accept import RecordToRecordTravel, SimulatedAnnealing
from alns.select import RouletteWheel, SegmentedRouletteWheel
from alns.stop import MaxIterations

alns = ALNS(rnd.RandomState(42))

alns.add_destroy_operator(random_destroy)  # Random Removal heuristic
alns.add_destroy_operator(worst_destroy)  # Worst Removal heuristic
alns.add_destroy_operator(static_rel_destroy)  # Static-relative Removal heuristic

alns.add_repair_operator(greedy_insertion)  # Greedy Insertion heuristic
alns.add_repair_operator(regret_insertion)  # Regret-2 Insertion heuristic
alns.add_repair_operator(regret_repair)  # Regret-3 Insertion heuristic
```

```python
def two_opt_star(solution):
    """
    Implementation of 2-opt* operator for multiple routes.
    """
    routes = solution.routes
    best_route = None
    best_cost = None
    improvement = True
    while improvement:
        improvement = False
        for p, route1 in enumerate(routes):
            for q, route2 in enumerate(routes):
                if p == q:
                    for i in range(1, len(route1) - 1):
                        for j in range(i + 1, len(route2)):
                            new_route = route1[:i] + route1[i:j][::-1] + route2[j:]
                            new_cost = route_cost(route1) - route_cost(new_route)
                            if best_cost is None or new_cost > best_cost:
                                best_cost = new_cost
                                best_route = (p, new_route)
                                improvement = True
                else:
                    for i in range(1, len(route1) - 1):
                        for j in range(1, len(route2)):
                            new_route1 = route1[:i] + route2[j:]
                            new_route2 = route2[:j] + route1[i:]
                            new_cost = route_cost(route1) - route_cost(new_route1) + route_cost(route2) - route_cost(new_route2)
                            if best_cost is None or new_cost > best_cost:
                                best_cost = new_cost
                                best_route = (p, new_route1, q, new_route2)
                                improvement = True
        routes[best_route[0]] = best_route[1]
        if len(best_route) == 4:
            routes[best_route[2]] = best_route[3]
```

```python
class Iterations:
    def __name__(self):
        return "Iterations"
    def __init__(self):
        self.iter = 0
    def __call__(self, sol, rnd):
        self.iter += 1
        if self.iter % 100 == 0:
            two_opt_star(sol)
        print(self.iter, sol.cost)

iterations = Iterations()
```

```python
init = init_sol()
select = SegmentedRouletteWheel([25, 5, 1, 0], 0.8, 5, 3, 3)
accept = RecordToRecordTravel.autofit(init.objective(), 0.2, 0, 30)
stop = MaxIterations(40)

alns.on_accept(iterations)
alns.on_reject(iterations)
result = alns.iterate(init, select, accept, stop)
```
